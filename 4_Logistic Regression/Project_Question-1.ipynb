```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression

import pandas as pd
```

```python
titanic= pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")

df= titanic.copy()
```

```python
df.isnull().sum()
```

### It is clearly visible that age has a lot of null entries and since the no is very high therefore we will
### fill the null entries with some value istead of simply dropping it. 

```python
df.Age.fillna(df.Age.mean(), inplace= True)

df
```

```python
def func( str ):
    if( str == "female" ):
        return 0
    else:
        return 1
```

```python
df["Sex"]= df.Sex.apply(func)

df
```

```python
del df["Name"]
```

```python
df.drop(["Fare", "Cabin", "Ticket", "Embarked"], axis= 1, inplace= True)
```

```python
df
```

```python
df

X= df.iloc[ : , 0: 5 ]
Y= df.iloc[ : , 5 ]

print(X)
print(Y)
```

```python
algo1= LogisticRegression(C=2, solver='saga', tol= 0.001, multi_class='multinomial', penalty= 'l2', max_iter= 1000 )

algo1.fit(X, Y)
```

### Checking training score. It should not be 100% or 90% because this indicates that the algorithm is trying 
### remembering the training data. 

```python
algo1.score(X, Y)
```

```python
testData= pd.read_csv("0000000000002429_test_titanic_x_test.csv")

test_df= testData.copy()

test_df
```

```python
test_df.drop(["Name", "Ticket", "Fare", "Cabin", "Embarked"], axis= 1, inplace= True)
```

```python
test_df.Age.fillna(test_df.Age.mean(), inplace= True)
test_df


test_df["Sex"]= test_df.Sex.apply(func)

test_df
```

```python
X_Test= test_df.iloc[ : , 0 : 6 ]
```

```python
Y_pred= algo1.predict(X_Test)
```
